```python
meijer_df  = temp_df[temp_df['Mrkt Desc'] == 'Meijer Total TA']
meijer_df_reduced = meijer_df[['Fisc Pd', 'Fisc Yr', 'Mrkt Desc', 'YrPd', 'YYYYPP', '% Unit Non-Promoted', '% Unit Promoted ', 'Avg Prc', 'Avg. Price', 'Avg. Price yago', 'Data Update Date', 'DOL', 'Non-Promo Price', 'Non-Promo Price yago', 'Promo Price', 'Promo Price yago', 'TotalCnt', 'Units', 'Yago Dol', 'Yago Units']].copy()
```

```python
meijer_df_reduced = meijer_df_reduced.dropna(subset = ['Promo Price'])
meijer_df_reduced.reset_index(drop = True, inplace = True)
```

```python
meijer_df_reduced['% Unit Non-Promoted'] = meijer_df_reduced['% Unit Non-Promoted'].str.rstrip('%').astype(float) / 100
meijer_df_reduced['% Unit Promoted '] = meijer_df_reduced['% Unit Promoted '].str.rstrip('%').astype(float) / 100
```

```python
meijer_df_reduced
```

```python
meijer_df_reduced['Promo Sales'] = meijer_df_reduced['% Unit Non-Promoted'] * meijer_df_reduced['DOL']
meijer_df_reduced['Non Promo Sales'] = meijer_df_reduced['% Unit Promoted '] * meijer_df_reduced['DOL']
```

```python
meijer_df_reduced
```

```python
meijer_df_reduced['Promo Price'] = meijer_df_reduced['Promo Price'].str.replace('$', '')
meijer_df_reduced['Non-Promo Price'] = meijer_df_reduced['Non-Promo Price'].str.replace('$', '')
meijer_df_reduced['Avg. Price'] = meijer_df_reduced['Avg. Price'].str.replace('$', '')
```

```python
meijer_df_reduced = meijer_df_reduced.sort_values(by = 'YYYYPP')
meijer_df_reduced.reset_index(drop = True, inplace = True)
```

```python
meijer_df_reduced
```

```python
meijer_df_reduced['Days'] = [30, 31, 
                             31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31,
                             31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31,
                             31, 28, 31, 30, 31, 30, 31, 31, 30, 31]
```

```python
meijer_df_reduced['Promo Sales per Day'] = 
meijer_df_reduced['Non Promo Sales per Day']
```

```python
sum_vals_promo = meijer_df_reduced.groupby('Promo Price')['Promo Sales'].sum().reset_index()
sum_vals_nonpromo = meijer_df_reduced.groupby('Non-Promo Price')['Non Promo Sales'].sum().reset_index()

sum_vals_promo = sum_vals_promo.rename(columns={'Promo Price': 'Price'})
sum_vals_nonpromo = sum_vals_nonpromo.rename(columns={'Non-Promo Price': 'Price'})
```

```python
sum_vals_nonpromo
```

```python
meijer_poptarts8ct_salesvolume = pd.merge(sum_vals_promo, sum_vals_nonpromo, left_on = 'Price', right_on = 'Price', how = 'outer')
```
