```python
dict = {'FRSTD STRWBRY' : 'FRSTD STRWBRRY',
        'FRSTDSTRWBRRY' : 'FRSTD STRWBRRY',
        'FRSTD CHCFDG' : 'FRSTD CH FDG',
        'FRSTD BNSGRCIN' : 'FRSTD BRN SGR CIN',
        'FRSTD BRNSGRCIN' : 'FRSTD BRN SGR CIN'}
```

```python
pln['Mixd Plt Matrl Desc'] = pln['Mixd Plt Matrl Desc'].str.replace('FRSTD STRWBRY','FRSTD STRWBRRY')
pln['Mixd Plt Matrl Desc'] = pln['Mixd Plt Matrl Desc'].str.replace('FRSTDSTRWBRRY','FRSTD STRWBRRY')
pln['Mixd Plt Matrl Desc'] = pln['Mixd Plt Matrl Desc'].str.replace('FRSTD CHCFDG','FRSTD CH FDG')
pln['Mixd Plt Matrl Desc'] = pln['Mixd Plt Matrl Desc'].str.replace('FRSTD BNSGRCIN','FRSTD BRN SGR CIN')
pln['Mixd Plt Matrl Desc'] = pln['Mixd Plt Matrl Desc'].str.replace('FRSTD BRNSGRCIN','FRSTD BRN SGR CIN')
```

```python
sorted(pln['Mixd Plt Matrl Desc'].unique())
```

```python
sums1 = pln.groupby('Mixd Plt Matrl Desc')['Totl Pln Kmf Cost Of Prod Usd'].sum().reset_index()
sums2 = pln.groupby('Mixd Plt Matrl Desc')['Actl Cost Of Prod Usd'].sum().reset_index()
sums3 = pln.groupby('Mixd Plt Matrl Desc')['Actl Ship Usd'].sum().reset_index()
```

```python
final_df = pd.merge(sums1, sums2, on = 'Mixd Plt Matrl Desc')
```

```python
final_df = pd.merge(final_df, sums3, on = 'Mixd Plt Matrl Desc')
```
