```python
import pandas as pd
import imageio
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
poptarts = ['strawberry', '']
```

```python
strawberry = imageio.v3.imread('poptart_images/strawberry.webp')
```

```python
plt.figure(figsize = (15,15))
plt.axis('off')
plt.imshow(strawberry)
```

```python
plt.imshow(strawberry[:, :, 0])
```

```python
image_height, image_width, _ = strawberry.shape

# Reshape the image array to a list of (x, y, r, g, b) tuples
data = [(x, y, *strawberry[y, x]) for y in range(image_height) for x in range(image_width)]

# Create a DataFrame
strawberry_df = pd.DataFrame(data, columns=['x', 'y', 'R', 'G', 'B'])

# Display the DataFrame
strawberry_df
```
